```python
import random
import copy
import math
import numpy as np
import array as arr
```

```python
crossover_prob = 0.6
mutation_prob = 0.03
```

```python
#this function calculates number of attacking pairs
def fitness(individual):
    n = len(individual)
    rowTally = [0 for _ in range(n)]
    mDiagTally = [0 for _ in range(2*n-1)]
    sDiagTally = [0 for _ in range(2*n-1)]
    for i in range(n):
        rowTally[individual[i]-1] += 1
        mDiagTally[individual[i]-1 + i] +=1
        sDiagTally[n-individual[i]-1+i] += 1
    conflicts = 0
    for i in range(2*n-1):
        rowConflicts = 0
        mDiagConflicts = 0
        sDiagConflicts = 0
        if i < n:
            x = rowTally[i]
        y = mDiagTally[i]
        z = sDiagTally[i]
        conflicts += (x*(x-1))/2
        conflicts += (y*(y-1))/2
        conflicts += (z*(z-1))/2
    return conflicts        
```

```python
def crossover(individual1, individual2):
    randNo = random.random()
    if randNo > crossover_prob:
        return individual1
    n = len(individual1)
    rand = random.randint(0,n-1)
    for i in range(rand, n):
        individual1[i], individual2[i] = individual2[i], individual1[i]
    return individual1
    
```

```python
def mutation(individual):
    randNo = random.random()
    if randNo > mutation_prob:
        return individual
    n = len(individual)
    rand1 = random.randint(0,n-1)
    rand2 = random.randint(0,n-1)
    individual[rand1] = rand2
    return individual
```

```python
def generate_individual(n):
    result = list(range(0, n))
    np.random.shuffle(result)
    return result
```

```python
class Genetic(object):

    def __init__(self, n ,pop_size):
        #initializing a random individuals with size of initial population entered by user
        self.queens = []
        self.temp = []
        for i in range(pop_size):
            self.queens.append(generate_individual(n))

    #generating individuals for a single iteration of algorithm
    def generate_population(self, random_selections):
        while True:
            candid_parents = []
            candid_fitness = []
            #getting individuals from queens randomly for an iteration
            for i in range(random_selections):
                candid_parents.append(self.queens[random.randint(0, len(self.queens) - 1)])
                candid_fitness.append(fitness(candid_parents[i]))
            fitness_copy = copy.deepcopy(candid_fitness)
            parents_copy = copy.deepcopy(candid_parents)
            #sort the fitnesses of individuals
            parents_fitness = list(zip(parents_copy, fitness_copy))
            sorted_parents_fitness = sorted(parents_fitness, key = lambda x : x[1])
            #getting 2 first individuals(min attackings)
            parent1 = sorted_parents_fitness[0][0]
            parent2 = sorted_parents_fitness[1][0]
            #crossover the two parents
            child = crossover(parent1, parent2)
            mutated_child = mutation(child)
            #in code below check if each child is better than each one of queens individuals, set that individual the new child
            if len(self.temp) == len(self.queens):
                self.queens = copy.deepcopy(self.temp)
                self.temp = []
                #print("new generation ready")
                break
            if fitness(child) >= fitness(parent1):
                self.temp.append(mutated_child)
            

    def finished(self):
        best_fitness = 100
        solution = None
        foundSolution = False
        for i in self.queens:
            #we check if for each queen there is no attacking(cause this algorithm should work for n queen,
            # it was easier to use attacking pairs for fitness instead of non-attacking)
            fitnessLevel = fitness(i)
            if fitnessLevel == 0:
                solution = i
                foundSolution = True
            if fitnessLevel < best_fitness:
                best_fitness = fitnessLevel
        print(best_fitness)
        return [foundSolution, solution]
            
            

    def start(self, random_selections=3):
        #generate new population and start algorithm until number of attacking pairs is zero
        while not self.finished()[0]:
            self.generate_population(random_selections)
        final_state = self.finished()
        if final_state[0] == True:
            print(('Solution : ' + str([i+1 for i in final_state[1]])))
        else:
            print("Solution not found")
```

```python
# ******************** N-Queen Problem With GA Algorithm *********************** #
n=(int)(input('Enter the value of N \n -'))
initial_population=(int)(input('Enter initial population size \n -'))
algorithm = Genetic(n=n,pop_size=initial_population)
algorithm.start()
```
